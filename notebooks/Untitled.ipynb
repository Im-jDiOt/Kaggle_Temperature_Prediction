```python
for k in train_df.columns:
    # count_of_minus_9999 = train_df[train_df[k] == -9999][k].count()
    count_of_nan = train_df[train_df[k].isnull()][k].sum()
    print(f"{k}: {(count_of_nan)}")
```

```python
anomaly_ratios = {}

for k in train_df.columns:
    count_of_minus_9999 = train_df[train_df[k] == -9999][k].count()
    count_of_nan = train_df[train_df[k].isnull()][k].count()
    total_anomalies = count_of_minus_9999 + count_of_nan
    proportion = total_anomalies / train_df.shape[0]
    anomaly_ratios[k] = proportion

print("비율이 0이 아닌 이상치 비율:")
for key, value in anomaly_ratios.items():
    if value != 0:
        print(f"{key}: {value}")
```
