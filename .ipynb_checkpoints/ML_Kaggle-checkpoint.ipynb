```python
# Import 라이브러리 및 데이터 불러오기
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 데이터 경로에 맞게 데이터 불러오기
train_dataset = pd.read_csv('challenge-data/train_dataset.csv')
station_info = pd.read_csv('challenge-data/station_info.csv')

# 데이터 기본 정보 확인
print("Train Dataset Shape:", train_dataset.shape)
print("Station Info Shape:", station_info.shape)

print("\nTrain Dataset Columns:")
print(train_dataset.columns.tolist())

# 데이터 샘플 확인 (앞부분 5줄)
display(train_dataset.head())
display(station_info.head())
```

```python
# 데이터의 기본 정보 확인: 데이터 타입, 결측치 등
train_dataset.info()

# 기초 통계량 출력
display(train_dataset.describe())

# 타겟 변수(target)의 분포 시각화
plt.figure(figsize=(8, 5))
plt.hist(train_dataset['target'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Target (Temperature anomaly)')
plt.ylabel('Frequency')
plt.title('Distribution of Target Variable')
plt.show()
```

```python
import numpy as np

# -9999 값을 np.nan으로 교체하기 전 각 컬럼별 -9999 개수 확인 (예시: 일부 컬럼 확인)
cols_to_check = ['cloud_cover_0', 'dew_point_0', 'humidity_0']
for col in cols_to_check:
    count_neg = (train_dataset[col] == -9999).sum()
    print(f"{col}의 -9999 개수: {count_neg}")

# 전체 데이터에서 -9999를 NaN으로 대체
train_dataset.replace(-9999, np.nan, inplace=True)

# 컬럼별 결측치 비율 계산 (몇몇 주요 변수만 확인)
missing_ratios = (train_dataset.isnull().sum() / len(train_dataset)) * 100
missing_ratios = missing_ratios[missing_ratios > 0].sort_values(ascending=False)
print("결측치 비율이 있는 컬럼 (상위 10개):")
print(missing_ratios.head(10))

# 결측치 비율 시각화 (상위 20개 컬럼)
plt.figure(figsize=(10, 6))
missing_ratios.head(20).plot(kind='bar')
plt.ylabel('Missing Value Percentage (%)')
plt.title('Top 20 Missing Value Ratios in Train Dataset')
plt.show()
```

```python
# 시간대별 변수에 대한 기본 탐색: 예를 들어 'cloud_cover' 변수 (cloud_cover_0 ~ cloud_cover_23)
# 24시간 데이터를 각 시간대별 평균값으로 집계
cloud_cols = [col for col in train_dataset.columns if col.startswith('cloud_cover_')]
cloud_means = train_dataset[cloud_cols].mean()

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

plt.figure(figsize=(10, 6))
plt.plot(range(24), cloud_means.values, marker='o')
plt.xlabel('Hour')
plt.ylabel('Mean Cloud Cover')
plt.title('24시간별 평균 구름량')
plt.xticks(range(24))
plt.grid(True)
plt.show()

# 다른 예시: dew_point 변수의 시간대별 평균
dew_cols = [col for col in train_dataset.columns if col.startswith('dew_point_')]
dew_means = train_dataset[dew_cols].mean()

plt.figure(figsize=(10, 6))
plt.plot(range(24), dew_means.values, marker='o', color='orange')
plt.xlabel('Hour')
plt.ylabel('Mean Dew Point (°C)')
plt.title('24시간별 평균 이슬점 온도')
plt.xticks(range(24))
plt.grid(True)
plt.show()
```

```python
# %% [code]
# 1) 결측치 비율 계산 및 40% 이상 변수 제거
missing_ratio = train_dataset.isnull().sum() / len(train_dataset)
cols_to_drop = missing_ratio[missing_ratio > 0.4].index.tolist()
print("제거할 컬럼 수:", len(cols_to_drop))
print("제거할 컬럼 예시:", cols_to_drop[:5])

# 원본 데이터에 영향이 없도록 copy() 사용
data_clean = train_dataset.copy()
data_clean.drop(columns=cols_to_drop, inplace=True)

# 2) 나머지 결측치 중앙값으로 채우기
for col in data_clean.columns:
    if data_clean[col].isnull().any():
        median_val = data_clean[col].median()
        data_clean[col].fillna(median_val, inplace=True)

print("남은 결측치 개수:", data_clean.isnull().sum().sum())

# %% [markdown]
# **이상치 처리 (IQR 기반)**  
# 이상치의 경우, 각 변수별로 IQR을 구한 후, [Q1 - 1.5*IQR, Q3 + 1.5*IQR] 범위를 벗어나는 값을 한 번 확인하고, 필요 시 로그 변환이나 윈저화(Winsorization) 적용을 고려할 수 있음.
```

```python
# %% [code]
# 간단한 IQR 기반 이상치 탐지 (예시: 몇몇 주요 변수만 확인)
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] < lower_bound) | (df[col] > upper_bound)]

outlier_cols = ['target', 'climatology_temp']  # 예시로 타겟 변수와 기준 기온 점검
for col in outlier_cols:
    outliers = detect_outliers(data_clean, col)
    print(f"{col} 이상치 개수:", len(outliers))
```

```python
# %% [code]
import numpy as np

# 시간대별 집계: 예를 들어, cloud_cover에 대해 (0~5시: 새벽, 6~11시: 오전, 12~17시: 오후, 18~23시: 저녁)
def aggregate_by_period(df, col_prefix, periods={'새벽': range(0,6), '오전': range(6,12), 
                                                  '오후': range(12,18), '저녁': range(18,24)}):
    agg_data = pd.DataFrame(index=df.index)
    for period, hours in periods.items():
        cols = [f"{col_prefix}_{h}" for h in hours if f"{col_prefix}_{h}" in df.columns]
        if cols:
            agg_data[f"{col_prefix}_{period}_mean"] = df[cols].mean(axis=1)
            agg_data[f"{col_prefix}_{period}_max"] = df[cols].max(axis=1)
            agg_data[f"{col_prefix}_{period}_min"] = df[cols].min(axis=1)
    return agg_data

# 예시로 cloud_cover와 dew_point에 대해 집계
cloud_agg = aggregate_by_period(data_clean, 'cloud_cover')
dew_agg = aggregate_by_period(data_clean, 'dew_point')

# 집계된 피처들을 원본 데이터에 병합
data_clean = pd.concat([data_clean, cloud_agg, dew_agg], axis=1)
print("집계 피처 추가 후 데이터 형태:", data_clean.shape)

# %% [markdown]
# **슬라이딩 윈도우 피처**  
# 이전 3일치 (또는 관측치) 평균, 최대, 최소 등의 통계값 추가  
# (데이터셋이 일별 관측값이라고 가정하고 date 컬럼을 datetime으로 변환한 후 슬라이딩 윈도우 적용)
```

```python
# %% [code]
# 먼저 date 컬럼 datetime 형식으로 변환
data_clean['date'] = pd.to_datetime(data_clean['date'], format='%m-%d')  # 월-일 형식이므로 연도 정보가 없다면 임의로 1900년 기준
# 만약 연도 정보가 없는 경우, 데이터셋 자체의 순서를 활용하거나 별도 처리 필요

# 정렬 (만약 날짜가 순차적으로 정렬되어 있지 않으면)
data_clean.sort_values('date', inplace=True)

# 슬라이딩 윈도우 (예시: 이전 3일 평균을 target의 피처로 추가)
window_size = 3
data_clean['target_3d_mean'] = data_clean['target'].rolling(window=window_size, min_periods=1).mean()

plt.figure(figsize=(10, 6))
plt.plot(data_clean['date'], data_clean['target_3d_mean'], label='3일 평균 Target')
plt.xlabel('Date')
plt.ylabel('3일 평균 Target (Anomaly)')
plt.title('Sliding Window: 3-day Mean of Target')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# %% [code]
# 집계 피처 중 하나 (예: cloud_cover_새벽_mean)의 분포 확인
plt.figure(figsize=(8,5))
plt.hist(data_clean['cloud_cover_새벽_mean'].dropna(), bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('Cloud Cover 새벽 Mean')
plt.ylabel('Frequency')
plt.title('Distribution of 새벽 Mean Cloud Cover')
plt.show()

# 여러 피처 간 관계 시각화 (예: target vs. dew_point_오전_mean)
if 'dew_point_오전_mean' in data_clean.columns:
    plt.figure(figsize=(8,5))
    plt.scatter(data_clean['dew_point_오전_mean'], data_clean['target'], alpha=0.3)
    plt.xlabel('Dew Point 오전 Mean (°C)')
    plt.ylabel('Target (Temperature anomaly)')
    plt.title('Scatter Plot: 오전 Mean Dew Point vs Target')
    plt.grid(True)
    plt.show()
```
